```python
print(df.describe(include="object"))
```

```python
import matplotlib.pyplot as plt

numeric_cols = df.select_dtypes(include="number").columns

for col in numeric_cols:
    plt.hist(df[col], bins=30, edgecolor="black")
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Between Nutrients")
plt.show()

sns.pairplot(df[["calories", "protein", "total_fat", "carbohydrates"]])
plt.show()
```

```python
if "category" in df.columns:
    category_stats = df.groupby("category").mean(numeric_only=True)
    print(category_stats)

    category_stats.plot(kind="bar", figsize=(12, 6))
    plt.title("Average Nutrients by Category")
    plt.ylabel("Average Value")
    plt.show()
```

```python
df["protein_per_100cal"] = (df["protein"] / df["calories"]) * 100

df["fat_protein_ratio"] = df["total_fat"] / df["protein"]

print(df[["item", "protein_per_100cal", "fat_protein_ratio"]].head())
```

```python
print(df.nlargest(5, "protein_per_100cal")[["item", "protein_per_100cal"]])

print(df.nsmallest(5, "sodium")[["item", "sodium"]])
```

```python
plt.hist(df["calories"], bins=30, edgecolor="black")
plt.title("Distribution of Calories")
plt.xlabel("Calories")
plt.ylabel("Frequency")
plt.show()

# Boxplot
sns.boxplot(x=df["calories"])
plt.title("Boxplot of Calories")
plt.show()
```

```python
# Scatter plot calories vs fat
plt.scatter(df["total_fat"], df["calories"], alpha=0.6)
plt.title("Calories vs Total Fat")
plt.xlabel("Total Fat (g)")
plt.ylabel("Calories")
plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Between Nutrients")
plt.show()
```

```python
if "category" in df.columns:
    sugar_by_cat = df.groupby("category")["sugars"].mean().sort_values()

    sugar_by_cat.plot(kind="bar", figsize=(10,5))
    plt.title("Average Sugar per Category")
    plt.ylabel("Average Sugar (g)")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Boxplots per nutrient
numeric_cols = ["calories", "total_fat", "protein", "carbohydrates", "sodium", "sugars"]

for col in numeric_cols:
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()

# Table of flagged outliers (z-score > 3)
from scipy.stats import zscore

z_scores = df[numeric_cols].apply(zscore)
outliers = df[(z_scores.abs() > 3).any(axis=1)]
print("Flagged outliers:")
print(outliers[["item"] + numeric_cols])
```

“From this analysis, we see that calories are strongly linked with fat, desserts and drinks are sugar-heavy, and a few items stand out as extreme outliers in sodium. If the goal is healthier eating, focusing on nutrient density (protein per calorie) and avoiding sodium-heavy items can guide better choices. A natural next step would be to cluster foods into groups or build a recommendation tool for balanced eating.”
